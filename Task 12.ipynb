### Exercise 1: Multiplication
#### *Create two arrays with different shapes (e.g., `(3,)` and `(3, 3)`) and perform element-wise multiplication using numpy broadcasting.*

```python
import numpy as np 

a=np.random.randint(1,10,(3,))
b=np.random.randint(1,10,(3,3))
print("Matrix A is: \n", a, "\n")
print("Matrix B is: \n", b, "\n")
print("Matrix A*B is: \n", a*b, "\n")
```

### Exercise 2: Addition
#### *Create an array with shape `(2, 3, 4)` and another array with shape `(3, 4)` and perform element-wise addition using numpy broadcasting.*

```python
a=np.random.randint(1,10,(2,3,4))
b=np.random.randint(1,10,(3,4))
print("Matrix A is: \n", a, "\n")
print("Matrix B is: \n", b, "\n")
print("Matrix A+B is: \n", a+b, "\n")
```

### Exercise 3: Division
#### *Create an array with shape `(2, 3)` and another array with shape `(1, 3)` and perform element-wise division using numpy broadcasting.*

```python
a=np.random.randint(1,10,(2,3))
b=np.random.randint(1,10,(1,3))
print("Matrix A is: \n", a, "\n")
print("Matrix B is: \n", b, "\n")
print("Matrix A/B is: \n", a/b, "\n")
```

### Exercise 4: Subtraction
#### *Create an array with shape `(3, 1)` and another array with shape `(3,)` and perform element-wise subtraction using numpy broadcasting.*

```python
a=np.random.randint(1,10,(3,1))
b=np.random.randint(1,10,(3,))
print("Matrix A is: \n", a, "\n")
print("Matrix B is: \n", b, "\n")
print("Matrix A-B is: \n", a-b, "\n")
```
