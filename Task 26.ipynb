### Understanding Optimizers

```python
# Exercise: Compare different optimizers on the IRIS dataset

import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam, RMSprop

# Load the IRIS dataset
iris = load_iris()
x = iris.data
y = iris.target

# Split the dataset into training and validation sets
x_train, x_val, y_train, y_val = train_test_split(
    x, y, test_size=0.2, random_state=42)

# Create a simple neural network model
model = Sequential([
    Dense(64, activation='relu', input_shape=(4,)),
    Dense(3, activation='softmax')
])

# Define a list of optimizers to compare
optimizers = [SGD(), Adam(), RMSprop()]

# Train the model using each optimizer and compare their performance
for optimizer in optimizers:
    # Compile the model with the optimizer
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    model.fit(x_train, y_train, batch_size=32,
              epochs=10, validation_data=(x_val, y_val))
```

### Last Layer Activations

```python
# Exercise: Compare different last-layer activation functions on the IRIS dataset

# Create a simple neural network model with different last-layer activation functions
model = Sequential([
    Dense(64, activation='relu', input_shape=(4,)),
    Dense(3, activation='softmax')
])

# Define a list of activation functions to compare
activations = ['softmax', 'sigmoid', 'tanh']

# Train the model using each activation function and compare their performance
for activation in activations:
    # Create the model with the specified activation function
    model = Sequential([
        Dense(64, activation='relu', input_shape=(4,)),
        Dense(3, activation=activation)
    ])

    # Compile and train the model
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    model.fit(x_train, y_train, batch_size=32,
              epochs=10, validation_data=(x_val, y_val))
```

### Loss Functions and Evaluation Metrics

```python
# Exercise: Compare different loss functions and evaluation metrics on the IRIS dataset

# Create a neural network model for classification
model = Sequential([
    Dense(64, activation='relu', input_shape=(4,)),
    Dense(3, activation='softmax')
])

# Define a list of loss functions and evaluation metrics to compare
loss_functions = ['mean_squared_error', 'sparse_categorical_crossentropy']
evaluation_metrics = ['accuracy', 'mae', 'mse']

# Train the model using each loss function and evaluation metric and compare their performance
for loss_func in loss_functions:
    for metric in evaluation_metrics:
        # Compile the model with the specified loss function and evaluation metric
        model.compile(optimizer='adam', loss=loss_func, metrics=[metric])

        # Train the model
        model.fit(x_train, y_train, batch_size=32,
                  epochs=10, validation_data=(x_val, y_val))
```
